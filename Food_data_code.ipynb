```python
data.drop(['식품이미지','열량 (kcal)','번호','식품군','Unnamed: 14'],inplace=True, axis=1)
```

```python
data=data.dropna(axis=0).reset_index(drop=True)
```

```python
data.info()
```

```python
data['식품이름']=data['식품이름'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣]"," ")
```

```python
data['식품이름'][2554]
```

```python
check2=''
check='a'
for i in range(1,len(data['식품이름'])):
    list_pop=[]
    if check2!=check:
        check=data['식품이름'][i-1]
    for j in range(len(data['식품이름'][i])):
        list_pop.append(data['식품이름'][i][j])
        food=''.join(list_pop)
        if check==food:
            data=data.drop(i,axis=0)
            i+=1
            check2=food
            break
        else:
            check2=food
```

```python
data=data.reset_index(drop=True)
```

```python
data['식품이름']=data['식품이름'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣]","")
```
